```
def normalize(M):
  Num = len(M.columns) # quantidade de colunas
  for i in range(Num):
    Max=M.iloc[:,[i]].max()
    Min=M.iloc[:,[i]].min()
    a = float((Max+Min))/2
    b = float((Max-Min))/2
    M.iloc[:,[i]]=M.iloc[:,[i]].apply(lambda x:(x-a)/b)
  return (M)
Carvao_novo4 = normalize(Carvao_novo)
print(Carvao_novo4.head())
```
