```python
# bibliotecas usadas
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
# para gerar o grafico no browser
%matplotlib inline
!git clone http://github.com/rodolfojbrandao/PHD
%cd PHD
```

```python
# funcao normatiza dados
def Normatiza(x):
    strings=list(x)
    for i in strings:
        max_x=x[i].max()
        min_x=x[i].min()
        a=(max_x+min_x)/2
        b=(max_x-min_x)/2
        x[i]=(x[i]-a)/b
    return x
```

```python
# funcao retorna os dados a forma original
def Original(x,x_old):
    strings=list(x)
    for i in strings:
        max_x=x_old[i].max()
        min_x=x_old[i].min()
        a=(max_x+min_x)/2
        b=(max_x-min_x)/2
        x[i]=x[i]*b+a
    return x
```

```python
def rede(i,j):
    clf = MLPRegressor(solver='lbfgs',activation='tanh',alpha=1e-5,
                       hidden_layer_sizes=(i, j),random_state=1)
    return clf
```

```python
# leitura dos dados
DAT=pd.read_csv("ReposeAngleANN.csv")
# separa os dados em normatizados e originais
DAT_OLD=[]
# Guarda os dados antes de serem normatizados, para que não sejam perdidos
DAT_OLD=DAT.copy()
DAT=Normatiza(DAT)
```

```python
# separa os dados em treino e teste
train=DAT.sample(frac=0.8,random_state=None) # poderia ser frac=0.5? sim ou nao?
test=DAT.drop(train.index)
# obtem as variaveis independentes (treino e teste)
x_train=train.iloc[:,[0,1,2,3,4,5,6,7]]
x_test=test.iloc[:,[0,1,2,3,4,5,6,7]]
X_OLD=DAT_OLD.iloc[:,[0,1,2,3,4,5,6,7]]
#print(x_test.head())
#print('================================================================')
#print(x_train.head())
#print('================================================================')
#print(X_OLD.head())
```

```python
# obtem as variaveis dependentes (treino e teste)
Index=[8,9]
y_train=train.iloc[:,Index]
y_test=test.iloc[:,Index]
Y_OLD=DAT_OLD.iloc[:,Index]
#print(y_test.head())
#print('================================================================')
#print(y_train.head())
#print('================================================================')
#print(Y_OLD.head())
```

```python
N=10
M=10
test=[]
train=[]
train=Original(y_train,Y_OLD)
test=Original(y_test,Y_OLD)
col_names=list(y_train)
for i in range(1,N):
    #via de regra usa apenas (N), porem como payton começa do 0 mas não podemos ter zero neur
    for j in range(1,M):
        clf=rede(i,j)
        #clf=MLPRegressor(solver='lbfgs',activation='tanh',alpha=1e-5,hidden_layer_sizes=(i, j),random_state=1)
        clf.fit(x_train, y_train)
        y_calc_train=clf.predict(x_train)
        y_calc_test=clf.predict(x_test)
        y_calc_train=pd.DataFrame(y_calc_train)
        y_calc_test=pd.DataFrame(y_calc_test)
        y_calc_train.columns = col_names
        y_calc_test.columns = col_names
        y_calc_train=Original(y_calc_train,Y_OLD)
        y_calc_test=Original(y_calc_test,Y_OLD)
        resist_obs_train=[]
        resist_calc_train=[]
        resist_calc_train=y_calc_train['Superior'].copy()
        resist_obs_train=train['Superior'].copy()
        resist_obs_test=[]
        resist_calc_test=[]
        resist_calc_test=y_calc_test['Superior'].copy()
        resist_obs_test=test['Superior'].copy()
        mse=mean_squared_error(resist_obs_test,resist_calc_test)
        print(f'number of hidden layers are {i} and {j}')
        print("MSE teste=",mse)
        R2=r2_score(resist_obs_test,resist_calc_test)
        print("R^2 teste=",R2)
        print("==================================")
```
