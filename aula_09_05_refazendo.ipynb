```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#pegando dados da pasta
DAT=pd.read_csv("Hydrofoil_ago_29_2018.csv")
print(DAT.head())
```

```
#selecionando alguns dados de DAT
#os dados selecionados são as entradas e as saídas da rede
Index=[2,3,4,5,6,7]
DAT_select=[]
DAT_select=DAT.iloc[:,Index].copy()
print(DAT_select.head())
```

```
#normalizar os dados selecionados
def normalize(x):
  #pega o nome das colunas
  strings=list(x)
  for i in strings:
    #para cada coluna faz a normalização
    max_x=x[i].max()
    min_x=x[i].min()
    a=(max_x+min_x)/2
    b=(max_x-min_x)/2
    x[i]=(x[i]-a)/b
  return x
DAT_select_n=normalize(DAT_select)
print(DAT_select_n.head())
```

```
#pega aleatoriamente 80% dos dados para treinamento
#random_state=1 especifica a semente de aleatoriedade
train=DAT_select_n.sample(frac=0.8,random_state=1)
print(train.head())
#pega o restante dos dados para o teste
test=DAT_select_n.drop(train.index)
print(test.head())
```

```
#pega dados de entrada da rede
#esses dados são necessários na etapa de treinamento e teste
Indice=[0,1]
train_x=[]
train_x=train.iloc[:,Indice].copy()
print(train_x.head())
test_x=[]
test_x=test.iloc[:,Indice].copy()
print(test_x.head())
```

```
#pega dados de saida da rede
#esses dados são necessários na etapa de treinamento e teste
Indice=[2,3,4,5]
train_y=[]
train_y=train.iloc[:,Indice].copy()
print(train_y.head())
test_y=[]
test_y=test.iloc[:,Indice].copy()
print(test_y.head())
```

```
#chamando biblioteca para treinamento
from sklearn.neural_network import MLPRegressor
clf=MLPRegressor(solver='lbfgs',activation='tanh',alpha=1e-5,hidden_layer_sizes=(9,8),random_state=1)
clf.fit(train_x,train_y)
```

```
#saidas preditas pela rede no treinamento e no teste
w_train=clf.predict(train_x)
w_test=clf.predict(test_x)
print(w_train)
print(w_test)
#transformando w_train em dataframe
w_train=pd.DataFrame(w_train)
#atribuindo a linha de nomes a w_train
col_names=list(train_y)
w_train.columns=col_names
print(w_train.head())
#transformando w_test em dataframe
w_test=pd.DataFrame(w_test)
col_names=list(train_y)
w_test.columns=col_names
print(w_test.head())
```

```
#pegando dos dados originais das saídas da redes
DAT=pd.read_csv("Hydrofoil_ago_29_2018.csv")
DAT_select_sai=[]
Indice=[4,5,6,7]
DAT_select_sai=DAT.iloc[:,Indice].copy()
print(DAT_select_sai.head())
```

```
#função para retornar os dados a escala original
def Original (x,x_old):
  strings=list(x)
  for i in strings: 
    max_x=x_old[i].max()
    min_x=x_old[i].min()
    a=(max_x+min_x)/2
    b=(max_x-min_x)/2
    x[i]=x[i]*b+a
  return x
w_train=Original(w_train,DAT_select_sai)
print(w_train.head())
w_test=Original(w_test,DAT_select_sai)
print(w_test.head())
test_y=Original(test_y,DAT_select_sai)
print(test_y.head())
```
